```python
class CalciumMonitor(SpikeMonitor):
    def __init__(self,source):
        SpikeMonitor.__init__(self,source)
        print(source)


N = 100
tau = 10*ms
eqs = '''
dv/dt = (2-v)/tau : 1
'''

G = NeuronGroup(N, eqs, threshold='v>1', reset='v=0', method='exact')
G.v = 'rand()'
c=CalciumMonitor(G)

run(50*ms)

plot(c.t/ms, c.i, '.k')
xlabel('Time (ms)')
ylabel('Neuron index');
```

```python
'''
https://github.com/HelmchenLab/CalciumSim/blob/master/modelCalcium/modelCalcium.m
https://github.com/HelmchenLab/CalciumSim/blob/master/modelCalcium/etc/spkTimes2Calcium.m
tauOn ... 10ms
% ampFast ... 8% tauFast ... 60ms
% ampSlow ... 3% tauSlow ... 800ms
x = 0:(1/frameRate):duration;
y = (1-(exp(-(x-spkT)./tauOn))).*...
    (ampFast.*exp(-(x-spkT)./tauFast))+(ampSlow.*exp(-(x-spkT)./tauSlow)); 
% y = (1-(exp(-(x-spkT)./tauOn))).*(ampFast*exp(-(x-spkT)./tauFast));
y(x<spkT) = 0;
y(isnan(y)) = 0;
'''
```

```python
def spkTimes2Calcium(spkTime,tauOn=10,ampFast=0.1,tauFast=1,ampSlow=0.01,tauSlow=10):
    x=np.arange(0,1000,0.1)
    y=(1-np.exp(-(x-spkTime)/tauOn))*(ampFast*(np.exp(-(x-spkTime)/tauFast)))+(ampSlow*np.exp(-(x-spkTime)/tauSlow))
    y[x<spkTime]=0
    return y

y=spkTimes2Calcium(spkTime=6,tauOn=10,ampFast=0.1,tauFast=1,ampSlow=0.01,tauSlow=10)
import matplotlib.pyplot as plt
plt.plot(y)
```

```python
def calc_sum_one_neuron(spike_train):
    sm=np.zeros(10000,)
    for spike in spike_train:
        print(spike)
        sm+=spkTimes2Calcium(float(spike))
        #print(sm[:100])
    return sm
sm=calc_sum_one_neuron(c.spike_trains()[50])
plt.plot(sm[:50])
```
